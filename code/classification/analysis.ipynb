```python
import matplotlib.pyplot as plt

# Define the data
padding_data = [
    ("Padding 64", 32, 0.705),
    ("Padding 128", 64, 0.499),
    ("Padding 256", 128, 0.514),
    ("Perfect Padding", 597, 0.377),
    ("Selective padding First 25", 7.80, 0.572),
    ("Selective Padding Top 10", 3.10, 0.503),
    ("Dummy 10", 10, 0.67),
    ("Dummy 20", 20, 0.624),
    ("Dummy 40", 40, 0.552)
]

# Sort the data by increasing overhead
# padding_data = [[x[0],x[2],x[1]] for x in padding_data]
padding_data.sort(key=lambda x: x[1])
# Extract the data into separate lists
names = [x[0] for x in padding_data]
overheads = [x[1] for x in padding_data]
precisions = [x[2] for x in padding_data]

# Plot the data
plt.plot(overheads, precisions, 'o-')

# Add labels to the plot
plt.title("Precision vs Overhead")
plt.xlabel("Overhead")
plt.ylabel("Precision")

# Add labels to each point
for name, overhead, precision in padding_data:
    if "Selective Padding Top 10" in name:
        plt.text(overhead, precision, name, ha='left', va='center',color='r')
    else:
        plt.text(overhead, precision, name, ha='left', va='center')
        

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define the data
padding_data = [
    ("Padding 64", 32, 0.705),
    ("Padding 128", 64, 0.499),
    ("Padding 256", 128, 0.514),
    ("Perfect Padding", 597, 0.377),
    ("Selective padding First 25", 7.80, 0.572),
    ("Selective Padding Top 10", 3.10, 0.503),
    ("Dummy 10", 10, 0.67),
    ("Dummy 20", 20, 0.624),
    ("Dummy 40", 40, 0.552)
]

# Sort the data by increasing overhead
padding_data.sort(key=lambda x: x[1])

# Extract the data into separate lists
names = [x[0] for x in padding_data]
overheads = [x[1] for x in padding_data]
precisions = [x[2] for x in padding_data]

# Set the style using seaborn
sns.set_style("whitegrid")

# Set the font size and style for the plot
plt.rcParams.update({'font.size': 12, 'font.family': 'serif'})

# Create the figure and axis objects
fig, ax = plt.subplots()

# Plot the data
sns.lineplot(x=overheads, y=precisions, marker='o', color='#4C72B0', ax=ax)

# Add labels to the plot
ax.set_title("Precision vs Overhead", fontweight='bold')
ax.set_xlabel("Overhead (%)", fontweight='bold')
ax.set_ylabel("Precision", fontweight='bold')

# Add labels to each point
for name, overhead, precision in padding_data:
    ax.text(overhead, precision, name, ha='left', va='center', fontweight='bold')

# Customize the axis ticks and labels
ax.set_xlim(left=0)
ax.set_xticks(range(0, 650, 50))
ax.set_yticks([0.4, 0.5, 0.6, 0.7])
ax.tick_params(axis='both', which='major', length=7, width=1, direction='out')
ax.tick_params(axis='both', which='minor', length=4, width=1, direction='out')
ax.xaxis.set_tick_params(pad=7)
ax.yaxis.set_tick_params(pad=7)

# Customize the legend
ax.legend(['Precision'], loc='lower left', frameon=True, fancybox=True, facecolor='white', edgecolor='black')

# Save the plot as a PDF file for submission
# plt.savefig('precision_vs_overhead.pdf', bbox_inches='tight', dpi=300)

# Show the plot
plt.show()
```
